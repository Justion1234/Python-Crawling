```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
#1. 네이버 금융사이트에서 코스피 Top100사이트의 정보를 요청
res = req.get("https://finance.naver.com/sise/sise_quant.naver")
```

```python
res.text
```

```python
#컴퓨터가 알아먹을 수 있는 언어로 바꾸기 위해서 beautifulsoup 사용

#2. 사용자가 받아온 데이터(text) 컴퓨터가 알아 먹을 수 있게 컴퓨터용 언어로 변형!
# html 언어로 변형 > bs라이브러리!
soup = bs(res.text, 'lxml')
```

```python
#3. soup 데이터에서 내가 필요한 요소를 찾아오기!
soup.select("a.title")
```
