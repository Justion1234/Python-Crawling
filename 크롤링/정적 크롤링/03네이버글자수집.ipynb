```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
res = req.get("http://www.naver.com")
```

```python
res.text
```

```python
soup = bs(res.text, "lxml")
```

```python
soup
```

```python
# soup이라는 데이터에서 메일, 카페, 블로그 같은 글자들만 수집해줘~ 가지고와줘~
naver = soup.select("a.nav")
```
